```python
#Importing Library
import pandas as pd
```

```python
#importing dataset
dataset=pd.read_csv("PrePlacement.csv")
```

```python
#viewing dataset
dataset
```

```python
def stdNBgraph(dataset):
        # Coverte to standard Normal Distribution
    import seaborn as sns
    mean=dataset.mean()
    std=dataset.std()

    values=[i for i in dataset]

    z_score=[((j-mean)/std) for j in values]

    sns.distplot(z_score,kde=True)

    sum(z_score)/len(z_score)
    #z_score.std()
```

```python
stdNBgraph(dataset["salary"])
```

Resolution :  By using Z-Score, we have converted the salary to standard normal distribution
