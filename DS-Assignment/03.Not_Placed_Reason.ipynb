```python
#Importing Library
import pandas as pd
```

```python
#importing dataset
dataset=pd.read_csv("Placement.csv")
```

```python
#viewing dataset
dataset
```

```python
#we use the loc function to filter the dataset for rows where 'status' is 'Not Placed'.
#then add specific columns that may provide insights into the reasons for non-placement.
#taking marks for review
reasons_for_non_placement = dataset.loc[dataset['status'] == 'Not Placed', ['ssc_p', 'hsc_p',"degree_p", 'etest_p', 'mba_p']]
```

```python
reasons_for_non_placement
```

```python
#Checking mean for all marks - non placed students
reasons_for_non_placement.mean()
```

```python
#we use the loc function to filter the dataset for rows where 'status' is 'Placed'.
#then add specific columns that may provide insights into the reasons for placement.
#taking marks for review
reasons_for_placement = dataset.loc[dataset['status'] == 'Placed', ['ssc_p', 'hsc_p',"degree_p", 'etest_p', 'mba_p']]
```

```python
#Checking mean for all marks - placed students
reasons_for_placement.mean()
```

Resolution :  By comparing mean value for marks of both "placed" and "not placed" students we found that the reason is low marks.
