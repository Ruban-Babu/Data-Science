```python
#importing matplotlib
import matplotlib.pyplot as plt
```

```python
# Scatter plot of etest_p vs salary
#Analyzing the scatter plot can provide insights into the relationship between "mba_p" and "salary" 
#help to identify outliers in the data.
plt.scatter(dataset['mba_p'], dataset['salary'])
plt.xlabel('mba_p')
plt.ylabel('salary')
plt.title('mba_p vs salary')
plt.show()
```
