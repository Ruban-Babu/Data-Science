```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("PrePlacement.csv")
```

```python
dataset
```

```python
#number of row and columns
print("The number of rows : ",dataset.shape[0])
print("The number of columns : ",dataset.shape[1])
```

# Univariate Anaylsis

```python
import seaborn as sns
sns.distplot(dataset["ssc_p"],hist=True,kde=True,kde_kws={"color":"red"})
```

# Bivariate Anaylsis

```python
sns.jointplot(x="ssc_p", y="salary", data=dataset)
plt.show()
```

```python
sns.jointplot(x="ssc_p", y="salary", data=dataset, kind="hex")
plt.show()
```

```python
#kde-kurnel density estimation
sns.jointplot(x="ssc_p", y="salary", data=dataset, kind="kde")
plt.show()
```

```python
#pair plot
sns.pairplot(dataset,hue="gender",diag_kind="kde",kind="scatter",palette="husl")
```

# For Categorical Plot

```python
# x = categorical data
# y = Numerical data
sns.stripplot(x="ssc_b",y="ssc_p",data=dataset)
plt.show()
```

```python
sns.stripplot(x="gender",y="ssc_p",data=dataset)
plt.show()
```

#swarmplot() method is used to draw a non-overlapping scatter plot where one of the variables is a categorical variable. this function is similar to the seaborn. stripplot() method but the points are adjusted so as to not overlap only on the categorical axis.

```python
sns.swarmplot(x="gender",y="ssc_p",data=dataset)
plt.show()
```
